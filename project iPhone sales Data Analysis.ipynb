# top 10 highest rated iphone on flip card in india

```python
higest_rated = df.sort_values(by = ['Star Rating'],
                             
                             ascending = False)
higest_rated = higest_rated.head(10)
print(higest_rated['Product Name'])
                             
```

# how many highest rated iphone on flip kard have

```python
iPhones = higest_rated['Product Name'].value_counts()
label = iPhones.index
counts  = higest_rated['Number Of Ratings']
figure = px.bar(higest_rated,x = label,y = counts,title = 'Number Of Rating Of highest Rated iPhones')
figure.show()
```

```python
iphones = higest_rated['Product Name'].value_counts()
label = iphones.index
counts = higest_rated['Number Of Reviews']
figure=px.bar(higest_rated,x = label, y = counts, title = 'number of reviews of highest rated iPhones' )
figure.show()
```

# what is the relationship between the sales price of iphones and the number of rating on falipkard

```python
import plotly.express as px

figure = px.scatter(
    data_frame=df,
    x='Number Of Ratings',
    y='Sale Price',
    size='Discount Percentage',
    trendline='ols',
    title="Relation between Sale Price and Number of Ratings"
)

figure.show()
```

# Relationship between discount percentage & Number of rating of iphone

```python
import plotly.express as px

figure = px.scatter(
    data_frame=df,
    x='Number Of Ratings',   # remove extra space
    y='Discount Percentage',
    size='Sale Price',
    trendline='ols',
    title='Relation between Discount Percentage and Number of Ratings'
)

figure.show()
```

```python
most_expansive = df.loc[df['Sale Price'].indxmax()]
least_expensive = df.loc[df['Sale Price'].idmin()]
print("Most Expansive Product :")
print(most_expansive)
print('\n Least Expensive Product')
print(least_expensive)
```
